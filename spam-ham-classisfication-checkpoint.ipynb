```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download stopwords from NLTK if you haven't already
nltk.download('stopwords')

# Initialize the stemmer and stopwords list
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_and_stem(text):
    # Remove non-alphabetical characters
    text = re.sub(r'[^A-Za-z\s]', '', text)
    
    # Split the text into words
    words = text.split()
    
    # Remove stopwords and apply stemming
    stemmed_words = [ps.stem(word) for word in words if word.lower() not in stop_words]
    
    return ' '.join(stemmed_words)


df1['v2'] = df1['v2'].apply(clean_and_stem)
```

#### These are too many features. Most of them won't be of any use. So I use a variance threshold to remove those features whose variance is less than 0.15 of maximum variance

```python


# Assuming tfidf_vectors and tfidf_vectorizer are already defined
# Convert the sparse TF-IDF matrix to a dense array for variance calculation
tfidf_array = tfidf_vectors.toarray()

# Calculate the variance of each feature
feature_variances = np.var(tfidf_array, axis=0)

# Determine the maximum variance
max_variance = np.max(feature_variances)

# Set the threshold at 50% of the maximum variance
threshold = 0.075 * max_variance

# Get indices of features that have variance above the threshold
important_features_idx = np.where(feature_variances > threshold)[0]

# Get the names of these important features
important_features = tfidf_vectorizer.get_feature_names_out()[important_features_idx]

# Select only the important features from the tfidf_vectors
tfidf_vectors_reduced = tfidf_vectors[:, important_features_idx]

# Convert the reduced TF-IDF matrix to a DataFrame
df_tfidf_reduced = pd.DataFrame(tfidf_vectors_reduced.toarray(), columns=important_features)

# Concatenate the reduced TF-IDF DataFrame with the cleaned original DataFrame df1
df1 = pd.concat([df1.reset_index(drop=True), df_tfidf_reduced.reset_index(drop=True)], axis=1)

# Reset the index of df1 to avoid any index mismatches
df1.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import xgboost as xgb
import lightgbm as lgb

# Assuming df1 contains the TF-IDF features and 'label' is the target column

X=df1.drop(columns=['v1', 'v2', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1)
y=df1['v1']

# Perform stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Initialize and train Random Forest model
rf_model = RandomForestClassifier(n_estimators=1000, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions with Random Forest
rf_predictions = rf_model.predict(X_test)

# Print classification report and accuracy for Random Forest
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_predictions))
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_predictions)}")

# Initialize and train XGBoost model
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb_model.fit(X_train, y_train)

# Make predictions with XGBoost
xgb_predictions = xgb_model.predict(X_test)

# Print classification report and accuracy for XGBoost
print("XGBoost Classification Report:")
print(classification_report(y_test, xgb_predictions))
print(f"XGBoost Accuracy: {accuracy_score(y_test, xgb_predictions)}")

# Initialize and train LightGBM model
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(X_train, y_train)

# Make predictions with LightGBM
lgb_predictions = lgb_model.predict(X_test)

# Print classification report and accuracy for LightGBM
print("LightGBM Classification Report:")
print(classification_report(y_test, lgb_predictions))
print(f"LightGBM Accuracy: {accuracy_score(y_test, lgb_predictions)}")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df1 contains the TF-IDF features and 'label' is the target column


# Perform stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Initialize and train the SVM model
svm_model = SVC(probability=True, kernel='linear')  # Using a linear kernel for text classification
svm_model.fit(X_train, y_train)

# Make predictions
svm_predictions = svm_model.predict(X_test)

# Evaluate the model
print("SVM Classification Report:")
print(classification_report(y_test, svm_predictions))
print(f"SVM Accuracy: {accuracy_score(y_test, svm_predictions)}")

# Confusion Matrix
cm = confusion_matrix(y_test, svm_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for SVM Classifier')
plt.show()

# ROC Curve and AUC
y_proba = svm_model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = roc_auc_score(y_test, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'SVM (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for SVM Classifier')
plt.legend(loc='lower right')
plt.show()
```
